```python
import pandas as pd
```

```python
df = pd.read_csv('carros.csv', index_col=0)
display(df)
```

```python
primeiros_cinco = df.iloc[0:5]
display(primeiros_cinco)
```

```python
carro_b = df.loc["B"]
display(carro_b)
```

```python
marcas_e_modelos = df.loc['A':'C', ['Marca', 'Modelo']]
display(marcas_e_modelos)
```
