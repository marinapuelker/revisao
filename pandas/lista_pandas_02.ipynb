```python
import pandas as pd

# Fazendo a importação do arquivo de cadastro
cadastro = pd.read_csv('cadastro.csv')

# Exibindo a tabela para confirmar a leitura
display(cadastro)
```

Parte 1: Seleções e Filtros

```python
#1. Selecionar apenas a coluna 'Nome':
print(cadastro['Nome'])
```

```python
#2. Selecionar as colunas 'Nome' e 'Email':
print(cadastro[['Nome', 'Email']])
```

```python
#3. Selecionar a linha da 'Carla Rocha' (Dica: utilize a indexação implícita .iloc ou filtre pelo nome):
# print(cadastro.iloc[2])
print(cadastro.loc[cadastro['Nome'] == 'Carla Rocha'])
```

```python
#4. Selecionar as duas primeiras linhas do DataFrame:
print(cadastro.iloc[0:2])
```

```python
#5. Selecionar as linhas onde o telefone começa com o código de área '(21)' 
# (Dica: pesquise sobre o método .str.startswith() ou use .str.contains()):
print(cadastro.loc[cadastro['Telefone'].str.startswith('(21)')])
```

```python
#6. Selecionar as linhas onde o email contém '@gmail.com':
print(cadastro.loc[cadastro['Email'].str.endswith('@gmail.com')])
```

```python
#7. Selecionar as linhas onde o campo 'Endereço' está vazio (ou seja, quando o Pandas marca como NaN na importação do arquivo.
#  Dica: use o método .isnull()):
print(cadastro.loc[cadastro['Endereço'].isnull()])
```

```python
#8. Selecionar o nome e o telefone de quem tem email do 'yahoo.com':
print(cadastro.loc[cadastro['Email'].str.endswith('@yahoo.com'), ['Nome', 'Telefone']])
```

```python
#9. Selecionar as linhas onde o código de área do telefone é '(11)' OU '(41)':
print(cadastro.loc[(cadastro['Telefone'].str.startswith('(21)')) | (cadastro['Telefone'].str.startswith('(41)'))])
```

```python
#10. Selecionar as linhas onde o endereço NÃO está vazio (Dica: use ~ ou .notnull()):
#print(cadastro.loc[~cadastro['Endereço'].isnull()])
print(cadastro.loc[cadastro['Endereço'].notnull()])
```

Parte 2: Criar e Alterar Colunas

```python
#1. Adicionar uma coluna 'País' com o valor 'Brasil' para todos os registros.
cadastro['Pais'] = 'Brasil'
print(cadastro)
```

```python
#2. Adicionar uma coluna 'Tem_Gmail' que seja True se o email da pessoa contiver '@gmail.com' e False caso contrário.
cadastro['Tem_Gmail'] = cadastro['Email'].str.contains('@gmail.com')
print(cadastro)
```

```python
#3. Criar uma nova coluna chamada 'DDD' contendo apenas o código de área do telefone (removendo os parênteses e o restante do número).
cadastro['DDD'] = cadastro['Telefone'].str.slice(0, 3)
print(cadastro)
```

```python
cadastro['Telefone'].str[1:3]
```

```python
#4. Remover o valor do DDD (incluindo os parênteses e o espaço) e deixar apenas o número na coluna 'Telefone'.
cadastro['Telefone'] = cadastro['Telefone'].str.slice(5, 16)
print(cadastro)
```

```python
#5. Modificar a coluna 'Nome' para que todos os nomes fiquem em letras maiúsculas.
cadastro['Nome'] = cadastro['Nome'].str.upper()
print(cadastro)
```

```python
#6. Adicionar uma coluna 'Email_Provedor' que contenha apenas o provedor de email (por exemplo, 'gmail.com', 'yahoo.com', 'hotmail.com').
#cadastro['Email_Provedor'] = cadastro['Email'].str.split('@').str[1]
display(cadastro)
```

```python
#7. Para os Endereços não preenchidos (que ficaram como NaN), preencher com o texto 'NÃO INFORMADO'.
cadastro['Endereço'] = cadastro['Endereço'].fillna('NÃO INFORMADO')
display(cadastro)
```
